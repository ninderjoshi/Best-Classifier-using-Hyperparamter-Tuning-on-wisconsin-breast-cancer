```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('CSVs/dataset.csv')
dataset.head()
```

```python
# dataframe.isna() returns dataframe whether it have any null value or not.
# isna().sum() is used to look for which column containing what number of null values.
dataset.isna().sum()
```

```python
#dropping unnamed : 32 column
df=dataset.dropna(axis=1)# axis=0 is used to do some operation over row while axis = 1 is used while dealing with columns
```

```python
classes = df['diagnosis'].unique()
```

```python
classes = ['Malagnant', 'Benign']
class0, class1 = list(df['diagnosis'].value_counts())
def autopct_format(values):
    def my_format(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{v:d}'.format(v=val)
    return my_format

plt.pie([class0,class1],labels=classes,autopct=autopct_format([class0,class1]),
        startangle=90,explode=(0.05, 0.05), colors=['lightskyblue','gold'])
plt.axis('equal')
plt.legend()
```
